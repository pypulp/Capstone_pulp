```python
from pandas_datareader import wb
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# GDP per capita (constant 2015 USD)

is gross domestic product divided by midyear population. GDP is the sum of gross value added by all resident producers in the economy plus any product taxes and minus any subsidies not included in the value of the products. It is calculated without making deductions for depreciation of fabricated assets or for depletion and degradation of natural resources. Data are in constant 2015 U.S. dollars.

ID: NY.GDP.PCAP.KD

https://data.worldbank.org/indicator/NY.GDP.PCAP.KD

```python
data= wb.download(indicator='NY.GDP.PCAP.KD', country=['US','CN', 'IN','BR','RU'], start=2000, end=2021)
```

```python
data= data.unstack()
```

```python
data
```

```python
data=data.xs('NY.GDP.PCAP.KD', axis=1, drop_level=True)
```

```python
#data.columns=map(str,list(range(2000,2021,1)))
data.columns=map(int,list(range(2000,2021,1)))
```

```python
data.columns
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,4),dpi =200)
sns.lineplot(data =data.transpose(),marker ="v")
plt.legend(loc=(1.03,0.7))
plt.xlim(2000,2021)
plt.suptitle("GDP per capita (constant 2015 US$)")
plt.xticks(data.columns,rotation=45);
plt.savefig("GDP per capita (constant 2015 US$)1.jpeg",bbox_inches='tight')
```

# GDP growth (annual %)
Annual percentage growth rate of GDP at market prices based on constant local currency. Aggregates are based on constant 2015 prices, expressed in U.S. dollars. GDP is the sum of gross value added by all resident producers in the economy plus any product taxes and minus any subsidies not included in the value of the products. 

ID: NY.GDP.MKTP.KD.ZG

https://data.worldbank.org/indicator/NY.GDP.MKTP.KD.ZG?end=2020&start=2000

```python
gdpGrow= wb.download(indicator='NY.GDP.MKTP.KD.ZG', country=['US','CN', 'IN','BR','RU'], start=2000, end=2021)
```

```python
gdpGrow= gdpGrow.unstack()
gdpGrow
```

```python
gdpGrow=gdpGrow.xs('NY.GDP.MKTP.KD.ZG', axis=1, drop_level=True)
```

```python
gdpGrow.columns=map(int,list(range(2000,2021,1)))
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,7),dpi =200)
sns.lineplot(data =gdpGrow.transpose(),marker="X")
plt.legend(loc=(1.03,0.7))
plt.xlim(2000,2021)
plt.suptitle("GDP growth (annual %)")
plt.xticks(data.columns,rotation=45);
plt.savefig("GDP growth (annual %).jpeg",bbox_inches='tight')
```

# Population, total
Total population is based on the de facto definition of population, which counts all residents regardless of legal status or citizenship. The values shown are midyear estimates.

ID: SP.POP.TOTL

https://data.worldbank.org/indicator/SP.POP.TOTL

```python
pop= wb.download(indicator='SP.POP.TOTL', country=['US','CN', 'IN','BR','RU'], start=2000, end=2021)
```

```python
pop=pop.unstack()
```

```python
pop =pop.xs('SP.POP.TOTL', axis=1, drop_level=True)
```

```python
pop.columns=map(int,list(range(2000,2021,1)))
pop
```

```python
sns.set_style("darkgrid",{"grid.color": ".6", "grid.linestyle": ":"})
plt.figure(figsize=(12,7),dpi =200)
sns.lineplot(data =pop.transpose(), marker="o")
plt.legend(loc=(1.03,0.7))
plt.xlim(2000,2021)
plt.suptitle("Total Population")
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(data.columns,rotation=45);
plt.savefig("Total Population.jpeg",bbox_inches='tight')
```
